# Draw probability density functions of costs across random seeds
***

```python
%run -i imports.py
%matplotlib inline
```

```python
results_dir = '../results/'
```

```python
data_tags_list = ['H2-air', 'CH4-air', 'C2H4-air']
legend_labels = ['Hydrogen', 'Methane', 'Ethylene']
```

```python
power = 4
```

***

## Visualize the PDFs across different fuels

```python
color_map_colors = cmc.batlow
cluster_colors = color_map_colors(np.linspace(0, 1, 10))

n_components = 2

line_widths = [2,3,8]

fig = plt.figure(figsize=(5,4))

for i, data_tag in enumerate(data_tags_list):
    
    costs_without_Seta = pd.read_csv(results_dir + data_tag + '-costs-without-Seta-' + str(n_components) + 'D-LDM-power-' + str(power) + '.csv', header=None).to_numpy().ravel()
    costs_with_Seta = pd.read_csv(results_dir + data_tag + '-costs-with-Seta-' + str(n_components) + 'D-LDM-power-' + str(power) + '.csv', header=None).to_numpy().ravel()

    PDF_costs_without_Seta = pd.read_csv(results_dir + data_tag + '-PDF-without-Seta-' + str(n_components) + 'D-LDM-power-' + str(power) + '.csv', header=None).to_numpy().ravel()
    PDF_costs_with_Seta = pd.read_csv(results_dir + data_tag + '-PDF-with-Seta-' + str(n_components) + 'D-LDM-power-' + str(power) + '.csv', header=None).to_numpy().ravel()

    (_, idx_without) = preprocess.order_variables(np.array(costs_without_Seta)[None,:], method='mean', descending=True)
    if i==0: 
        plt.plot(np.array(costs_without_Seta)[idx_without], PDF_costs_without_Seta[idx_without], '--', lw=line_widths[i], c=cluster_colors[1], zorder=10)
    else:
        plt.plot(np.array(costs_without_Seta)[idx_without], PDF_costs_without_Seta[idx_without], '-', lw=line_widths[i], c=cluster_colors[1], zorder=10)
        
    (_, idx_with) = preprocess.order_variables(np.array(costs_with_Seta)[None,:], method='mean', descending=True)
    if i==0: 
        plt.plot(np.array(costs_with_Seta)[idx_with], PDF_costs_with_Seta[idx_with], '--', lw=line_widths[i], c=cluster_colors[7], label=legend_labels[i])
    else:
        plt.plot(np.array(costs_with_Seta)[idx_with], PDF_costs_with_Seta[idx_with], '-', lw=line_widths[i], c=cluster_colors[7], label=legend_labels[i])

plt.yticks([])

# if n_components == 3:
#     plt.xlim([0.9,3.1])
#     plt.xticks([1,2,3])
# if n_components == 2:
plt.xlim([0.25,11])
# plt.xticks([0,5,10])
plt.xlabel('$\mathcal{L}$', fontsize=26)
plt.ylabel('PDF', fontsize=20)
plt.xscale('log')
if n_components==2: plt.legend(ncol=1, frameon=False, fontsize=14)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
```
